```python
# Making a class for the Health Care System
class HIS:
    
    # Important details for the healthcare system like the doctors, nurses and patients are shown
    def __init__(self, doctors = None, nurses = None, patients = None):

        # If there are no doctors then an empty list of doctors is created
        if doctors is None:
            self.doctors = []
        else:
            self.doctors = doctors

        # If there are no nurses then an empty list of nurses is created
        if nurses is None:
            self.nurses = []
        else:
            self.nurses = nurses 

        # If there are no patients, then an empty list of patients is created
        if patients is None:
            self.patients = []
        else:
            self.patients = patients
        
    # This function creates a doctor and adds that doctor object to the list of doctors
    def create_doctor(self, doctor):
        self.doctors.append(doctor)
    
    # This function creates a nurse and adds that nurse object to the list of nurses
    def create_nurse(self, nurse):
        self.nurses.append(nurse)
    
    # This creates a new patient
    def create_patient(self, patient):
        self.patients.append(patient)

    # This shows all the patients
    def view_patients(self):
        for patient in self.patients:
            print(patient, '\n\n')
    
    # This shows all the doctors 
    def view_doctors(self):
        for doctor in self.doctors:
            print(doctor, '\n\n')

    # This shows all the nurses 
    def view_nurses(self):
        for nurse in self.nurses:
            print(nurse, '\n\n')
            
    # This assigns a doctor to the patient
    def assign_doctor(self, patient, doctor):
        for patient_stored in his.patients:
            if patient_stored.id == patient:
                patient_stored.doctor_id = doctor
    
    # This assigns a nurse to the patient
    def assign_nurse(self, patient, nurse):
        for patient_stored in his.patients:
            if patient_stored.id == patient:
                patient_stored.nurse_id = nurse
    
    # This assigns a medication to the patient
    def assign_medication(self, patient, medication, cost_of_medication):
        patient.medications.append(medication) 
        patient.bill += cost_of_medication

    # This assigns a lab test to the patient
    def assign_lab_test(self, patient, lab_test, cost_of_test):
        patient.lab_test = lab_test
        patient.bill += cost_of_test
    
    # This discharges a patient
    def discharge_patient(self, patient):
        self.patients.remove(patient)
        with open(f"{patient.first}{patient.last}_discharged", 'w') as f:
            f.write(f"Patient Name: {patient.first} {patient.last}\nPatient Bill: {patient.bill}$")

# Creating an object of the HIS class
his = HIS()
his.create_doctor(admin_doctor)
```

```python
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

cyan = bcolors.OKCYAN
pink = bcolors.HEADER
blue = bcolors.OKBLUE
green = bcolors.OKGREEN
red = bcolors.FAIL
end = bcolors.ENDC
bold = bcolors.BOLD
underline = bcolors.UNDERLINE


# Making a class for the Patient
class Patient:

    # This initializes the patient class with the required details
    def __init__(self, first, last, id, age, address, admit_date, height, weight, bp_sys, bp_dias, pulse, smoker, bill, doctor, nurse, medications = None, lab_test = None):
        self.first = first
        self.last = last
        self.id = id
        self.age = age
        self.address = address
        self.admit_date = admit_date
        self.height = height
        self.weight = weight
        self.bp_sys = float(bp_sys)
        self.bp_dias = float(bp_dias)
        self.pulse = float(pulse)
        self.smoker = smoker
        self.bill = float(bill)
        self.doctor_id = doctor
        self.nurse_id = nurse
        self.medications = medications
        self.lab_test = lab_test

        # If no medication is assigned then an empty list of medications is made
        if self.medications is None:
            self.medications = []
        else:
            self.medications = medications
        
        # If no lab test is assigned then an empty list of medications is made
        if self.lab_test is None:
            self.lab_test = []
        
        else:
            self.lab_test = lab_test

    # This function shows the information of a patient with the name masked
    def show_information(self):
        print(f"{bold}{green}Hello! Here are your details: {end}\n\n")
        print(f"First Name: {bold}{self.first[0]}***{end}")
        print(f"Last Name: {bold}{self.last[0]}***{end}")
        print(f"Patient ID: {bold}{self.id}{end}")
        print(f"Age: {bold}{self.age}{end}")
        print(f"Address: {bold}{self.address}{end}")
        print(f"Admit Date: {bold}{self.admit_date}{end}")
        print(f"Height: {bold}{self.height}{end}")
        print(f"Weight: {bold}{self.weight}{end}")
        if self.bp_sys >= 130 or self.bp_sys <= 90:
            print(f"{red}BP Systolic Last Checked: {self.bp_sys} mmHg {end}")
        else:
            print(f"{green}BP Systolic Last Checked: {self.bp_sys} mmHg{end}")
        if self.bp_dias >= 90 or self.bp_dias <= 60:
            print(f"{red}BP Diastolic Last Checked: {self.bp_dias} mmHg {end}")
        else:
            print(f"{green}BP Diastolic Last Checked: {self.bp_dias} mmHg{end}")
        if self.pulse >= 100 or self.pulse <= 60:
            print(f"{red}Pulse Last Checked: {self.pulse} BPM {end}")
        else:
            print(f"{green}Pulse Last Checked: {self.pulse} BPM {end}")
        if self.smoker.lower() == "yes":
            print(f"{red}Smoker: {self.smoker}{end}")
        else:
            print(f"{green}Smoker: {self.smoker} {end}")
        print(f"Total Expenses: {bold}{self.bill}{end}")
        for doctor in his.doctors:
            if doctor.id == self.doctor_id:
                print(f"{blue}Assigned Doctor ID: {doctor.id}\nDoctor Name: {doctor.first} {doctor.last}")
        
        for nurse in his.nurses:
            if nurse.id == self.nurse_id:
                print(f"{blue}Assigned Nurse ID: {nurse.id}\nNurse Name: {nurse.first} {nurse.last}{end}")
        print(f"Medications:\n")

        # If a person has atleast one medication prescribed, then the details will be printed out
        if len(self.medications) != 0:
            print(f"{bold}Medications assigned to Patient {green}{self.first} {self.last}: {end}")
            if len(self.medications) == 1:
                print(f"\tName of medication: {bold}{self.medications[0][0]}{end}")
                print(f"\tDosage: {bold}{self.medications[0][1]}{end}")
                print(f"\tRoute: {bold}{self.medications[0][2]}{end}")
                print(f"\tFrequency: {bold}{self.medications[0][3]}{end}\n\n")
            else:
                for medication in self.medications:
                    print(f"\tName of medication: {bold}{medication[0]}{end}")
                    print(f"\tDosage: {bold}{medication[1]}{end}")
                    print(f"\tRoute: {bold}{medication[2]}{end}")
                    print(f"\tFrequency: {bold}{medication[3]}{end}\n\n")
        # If a person has been prescribed no medication, then it says that no medication has been assigned
        else:
            print(f"{bold}\tNo medications assigned to patient {green}{self.first} {self.last}{end}")
        
        # If a person has at least one lab test assigned, then the details will be printed out
        if len(self.lab_test) != 0:
            print(f"{bold}Lab Tests assigned to Patient {green}{self.first} {self.last}: {end}")
            print(f"\tName of test: {bold}{self.lab_test[0]}{end}")
            print(f"\tTime of Test: {bold}{self.lab_test[1]}{end}\n\n")
        
        # If a person has no lab tests assigned then this will be printed out
        else:
            print(f"{bold}\tNo Lab tests assigned to patient {green}{self.first} {self.last} {end}")

# This returns the user friendly way to look at a patient object
    def __repr__(self):
        return f"{self.id}"
```
